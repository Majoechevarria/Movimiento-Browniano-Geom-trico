## Proceso de precios de acciones Netflix con Movimiento Browniano Geométrico

## Volatilidad
Las acciones tienen movimientos en su precio, suben y bajan. La volatilidad mide la desviación estándar del movimiento del precio de un activo financiero en un periodo de tiempo determinado. Al haber un incremento repentino en el precio se dice que ha aumentado la volatilidad.
La volatilidad en los mercados financieros, siempre se ve como una medida de riesgo del precio. Cuando un valor es más volátil, suele tener unos movimientos más bruscos que otros que no lo son tanto.



## Precio
El precio de la acción va a depender de distintos factores, como la oferta y demanda que a su vez esta altamente influenciada por la información financiera y noticias que puedan afectar o ayudar a la reputación de la empresa. El volumen de las acciones muchas veces ayuda a identificar estos eventos ya que precede al precio y después de encontrarnos con un periodo donde haya un cambio drástico en el volumen podemos esperar lo mismo en el precio. Por ejemplo, al aumentar la demanda de las acciones de Netflix eso se verá reflejado en periodos posteriores con un alza en el precio. Sin embargo, para este trabajo no consideraremos los factores anterior mencionados y nos basaremos en la volatilidad al ser una variable aleatoria para asi poder modelar los precios mediante un movimiento Browniano geométrico que definieremos más adelante. El cambio diario en los precios lo vamos a identificar con los rendimientos, que al ser positivos nos indicaran un incremento de un día respecto al anterior y al ser negativo nos indicara que el precio bajo en base al día anterior.


## Características del modelo

Un proceso aleatorio que describe el comportamiento de ciertas variables aleatorias a medida que se desplazan en el tiempo. Este proceso utiliza con frecuencia modelos financieros  para describir el comporamiento de los precios a largo plazo. Cuando se aplica a precios, el movimiento browniano tiene el supuesto que el cambio de un periodo de tiempo al siguiente no esta relacionado con los precios pasados. Es decir, cada cambio de precio es independiente de los cambios de precio anteriores y la volatilidad de los cambios de precios es constante.

El precio puede encontrarse en alguno de los siguientes 3 estados en un momento $n$ dado:
 1. Sube. 
 2. Se mantiene.
 3. Baja.

Un movimiento Browniano unidimensional de parametro $\sigma^{2}$ es un proceso estocastico {$B_{t}$ : t $\ge 0\$} con valores en R que cumple con las siguientes propiedades. 

1. B0 = 0
2. las trayectorias son continuas.
3. El proceso tiene incrementos independientes.
4. Para cualquier tiempo 0 $\le s < t$,  la variable incremento $B_{t} - B_{s}$ tiene distribucion N(0,$\sigma^{2}$(t-s)),

Movimiento browniano definido por la ecuación diferencial estocástica:
     
$S(t_{0}) = S_{0}$

$dS(t)= \mu S(t)dt +\sigma S(t)dW(t)$

$\frac{dS(t)}{S(t)}= \mu dt+\sigma dW(t)$

Donde:

1. Las trayectorias $t$ a $S(t)$ son continuas

2. Para todo $s < t$ el incremento relativo es independiente

3. La ley del cociente con parametros $\mu -\frac{1}{2}\sigma^{2}$)($t-s$),$\sigma^{2}$($t-s$)

Posteriormente a traves de las integrales estocasticas de Ito y cambios de probabilidad bajo el Teorema de Girsanov.

$S(T) = S_{0}e^{\int_{0}^{T}  \!(\mu S(t)dt - \frac{1}{2}\sigma^{2}S(t) )dt + \int_{0}^{T}  \! \sigma dW(t))}$

$S(t) = S_{0}e^{(\mu S(T)dt - \frac{1}{2}\sigma^{2}S(T) )dt +\sigma dW(T))}$, donde $T < t$

Donde:
1. Donde $W(t)$ obedece una ley Normal (0,1)

2. Valor esperado: $E[S(t_{0} + T)] = S_{0}e^{\mu T}$



$\delta S_{t}$ = $\mu S_{t} \delta t$ + $\sigma S_{t}$$\delta B_{t}$ 


$S_{t+1}$ = $S_{t}exp$(($\mu$ -$\frac{1}{2}\sigma^{2}$)$\delta t$ + $\sigma\delta B_{t}$) 

Donde:

$S$:Es el precio de la acción

$\mu$:los retornos esperados

$\sigma$:Es la volatilidad

$d B_{t}$: Es el diferencial del Mov. Browniano

## Simulación de precios de Netflix Inc. último trimestre de 2020

```python
import pandas as pd
from pandas_datareader import data as pdr
import numpy as np
import datetime as date
import matplotlib.pyplot as plt
from scipy import stats
import matplotlib.pyplot as plt
import plotly.express as px
import statsmodels.stats.stattools as sm
```

```python
fechafin = date.datetime(2020,9,30)
fechainicio = date.datetime(2017,1,1)

Nx = pdr.get_data_yahoo('NFLX', start = fechainicio, end = fechafin)['Close']
Precios = pdr.get_data_yahoo('NFLX', start = fechainicio, end = fechafin)
Precios = pd.DataFrame(Precios['Close'])
```

```python
Nx.plot()
```

```python
N = len(Precios.index)
i = 0
Rendimientos = []

for i in range(N-1):
    r = np.log(Precios.iloc[i+1,0]/Precios.iloc[i,0])
    Rendimientos.append(r)
    i = i+1
Rendimientos = pd.DataFrame(Rendimientos)
Rendimientos.columns = ['Rendimientos']
Rendimientos.index = Precios.index[1:N]
```

## Test de Skewness que la distribucion de los rendimientos sea simétrica

```python
plt.hist(Rendimientos)
```

```python
stats.mstats.skewtest(Rendimientos)
```

### debido a que el p-valor es $> 0.05$ se asume que los rendimientos se distribuyen de manera simétrica

## Prueba Durbin-Watson para comprobar independencia

```python
sm.durbin_watson(Rendimientos, axis=0)
```

```python
last_year = np.log(Nx[-252:]) #porcentajes logaritmicos
sigma = last_year.pct_change().std()*(252**0.5)#regla de la raiz cuadrada
```

```python
mu = (last_year[-1:].values/last_year[:1].values) - 1 #inicio y fin de ano para dar retornos esperados
```

```python
N = 66
T = 1.0
ult_precio = Nx[-1:]
```

```python
import scipy as sp
%matplotlib inline
```

```python
sp.random.seed(10)
paths = 7
dt = T/N
S = sp.zeros([N], dtype=float)#arreglo de 0
x = range(0, int(N),1)

df = pd.DataFrame()#almacenar trayectorias
#FIGURA
fig = plt.figure()
axis = fig.add_subplot(111)
for j in range(0, paths):
    S[0]= ult_precio
    for i in x[:-1]:
        e=sp.random.normal()
        S[i+1]=S[i]+S[i]*(mu-0.5*pow(sigma,2))*dt+sigma*S[i]*sp.sqrt(dt)*e
    df[j] = S
    plt.plot(x, S,lw=2)

plt.title('Caminos con Movimiento Browiniano')
axis.set_xlabel('Tiempo')
axis.set_ylabel('Precio')
axis.grid(True)
plt.show()
```

```python
df2 = pd.concat([Nx,df], axis = 0)
df2 = df2.reset_index()
df2 = df2.drop(['index'], axis = 1)
```

```python
plt.style.use('dark_background')
df3 = df2[-252:]
fig2 = plt.figure(figsize = (16,8))
axis2 = fig2.add_subplot(111)
plt.plot(df3)
plt.title('Simulacion Precio de Netflix con Movimiento Browniano')
axis2.set_xlabel('Tiempo')
axis2.grid(True)
plt.tick_params(
    axis='x',
    which='both',
    bottom=False,
    top=False,
    labelbottom=False)
plt.show()
```
